## Groupby Sales and Profit

```python
category_s=data.groupby('Category').Sales.sum().nlargest(n=100)
category_p=data.groupby('Category').Profit.sum().nlargest(n=100)
```

```python
print(category_s)
print('* * * * *')
print(category_p)
```

```python
data.head()
```

```python
data['Region'].unique()
```

## PROFIT ANALYSIS OF CATEGORY

```python
plt.style.use('seaborn')
category_s.plot(kind='bar',figsize=(12,6),fontsize =14,legend=True)
category_p.plot(kind='bar',figsize=(12,6),fontsize =14,color='red',legend=True)
plt.xlabel('Category',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title('Top Category Sales vs Profits',fontsize=15)
plt.show()
```

## PROFIT ANALYSIS OF SUB-CATEGORY

```python
category_s=data.groupby('Sub-Category').Sales.sum().nlargest(n=100)
category_p=data.groupby('Sub-Category').Profit.sum().nlargest(n=100)
plt.style.use('seaborn')
category_s.plot(kind='bar',figsize=(12,6),fontsize =14,legend=True)
category_p.plot(kind='bar',figsize=(12,6),fontsize =14,color='orange',legend=True)
plt.xlabel('Sub-Category',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title('Top Sub-Category Sales vs Profits',fontsize=15)
plt.show()
```

## PROFIT ANALYSIS OF REGION

```python
plt.figure(figsize=(5,5))
plt.title('Region')
plt.pie(data['Region'].value_counts(),labels=data['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

## Top state sales vs profit

```python
category_s=data.groupby('State').Sales.sum().nlargest(n=100)
category_p=data.groupby('State').Profit.sum().nlargest(n=100)
plt.style.use('seaborn')
category_s.plot(kind='bar',figsize=(12,6),fontsize =14,legend=True)
category_p.plot(kind='bar',figsize=(12,6),fontsize =14,color='red',legend=True)
plt.xlabel('States',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title('Top States Sales vs Profits',fontsize=15)
plt.show()
```
